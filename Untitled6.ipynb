```python
#import the rquired libary
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#import the Dataset
df = pd.read_csv("/content/random_stock_market_dataset.csv")
```

```python
df['Volume'] = df['Volume']/1000
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Day'] =df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
#Seperate the Close into closegroup
df['Close_group'] =pd.cut(df['Close'],bins=4)
df.head()
```

```python
df['Close_group'].value_counts()
```

Exploring each column using char

```python
# The first chart show the variation in stock market price at the open and close the market
df.plot.area(x='Date',y=['Open','Close'],alpha=0.5,figsize=(12,6))
plt.title('Varaiation in stock market price in 2024')
plt.xlabel("Date")
plt.ylabel('open and close price')
plt.grid(True,color='gray',linestyle = '--',linewidth=0.4,alpha=0.5)
plt.show()
```

```python
#The following chart shows the trading volume over the months.
df.plot(x='Date', y='Volume', color='k', marker='o', linewidth=1, linestyle='--', figsize=(12,6))
plt.title('Daily trading during the first two months of 2024')
plt.xlabel('Jan - Feb (2024)')
plt.ylabel('Trading volume (k)')
plt.grid(True, color='gray', linestyle='--', linewidth=0.4, alpha=0.4)
plt.show();
```

```python
#This last chart represents the trading volume of stocks in thousands (k) by category (Group_Close). In other words, what was the trading volume for shares between 120 and 216, and so on.
p1 = df.groupby('Close_group')['Volume'].sum()
p1.plot.bar(edgecolor='k', alpha=0.4, linewidth=1.4, figsize=(12,6))
plt.title('Volume in thousands of these shares sold over two months, by group price')
plt.xlabel('Price closing range')
plt.ylabel('Trading volume (k)')
plt.xticks(rotation=0)
plt.show()
```
